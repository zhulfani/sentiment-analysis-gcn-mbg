```python
import re
import string

def cleaningText(text):
    text = re.sub(r'@[A-Za-z0-9]+', ' ', text) # menghapus mention
    text = re.sub(r'#[A-Za-z0-9]+', ' ', text) # menghapus hashtag
    text = re.sub(r'RT[\s]', ' ', text) # menghapus RT
    text = re.sub(r"http\S+", ' ', text) # menghapus link
    text = re.sub(r'[0-9]+', ' ', text) # menghapus angka
    text = re.sub(r'[^\w\s]', ' ', text) # menghapus karakter selain huruf dan angka
    text = re.sub(r'[^\x00-\x7F]+', ' ', text)
    text = re.sub(r'(.)\1{2,}', r'\1', text) #menghapus duplikat huruf >= 3
    text = text.replace('\n', ' ') # mengganti baris baru dengan spasi
    text = text.translate(str.maketrans(' ', ' ', string.punctuation)) # menghapus semua tanda baca
    text = text.strip(' ') # menghapus karakter spasi dari kiri dan kanan teks
    return text
```

```python
def normalize_text(text, norm_dict):
    text = text.lower()
    words = text.split()
    normalized_words = [norm_dict.get(word, word) for word in words]
    return ' '.join(normalized_words)

df['normalized_text'] = df['text_clean'].astype(str).apply(lambda x: normalize_text(x, alay_dict))
```

```python
def normalize_tokens(tokens, norm_dict):
    """Normalize a list of tokens using the normalization dictionary"""
    if isinstance(tokens, list):
        # Process token by token if it's already a list
        return [norm_dict.get(token, token) for token in tokens]
    elif isinstance(tokens, str):
        # If for some reason it's still a string, try to make it work
        try:
            # Try to safely evaluate if it looks like a string representation of a list
            if tokens.startswith('[') and tokens.endswith(']'):
                # This is risky, but we're trying to recover from the situation
                import ast
                token_list = ast.literal_eval(tokens)
                return [norm_dict.get(token, token) for token in token_list]
            else:
                # Regular string normalization
                words = tokens.split()
                return [norm_dict.get(word, word) for word in words]
        except:
            # If evaluation fails, return original
            return tokens
    else:
        # If it's neither a list nor string, return as is
        return tokens

# Apply normalization correctly
df['token_normalized_text'] = df['normalized_text'].apply(lambda x: normalize_tokens(x, alay_dict))
```

```python
import nltk
from nltk.corpus import stopwords

# Download stopwords jika belum ada
try:
    nltk.data.find('corpora/stopwords')
except LookupError:
    nltk.download('stopwords')

# Ambil stopwords Bahasa Indonesia
stop_words = set(stopwords.words('indonesian'))

# Tambahkan stopwords tambahan
additional_stopwords = {'yang', 'ya', 'apa', 'ke', 'di', 'dari', 'adalah', 'kah', 'lah', 'pun'
                        # Kata sambung/partikel yang benar-benar tidak informatif
                        "nih", "sih", "kan", "tuh", "mah", "nya", "aja", "ya", "gitu", "oke", 
                        "si", "loh", "mulu", "ujungnya", "deh", "dong", "lagi", "udah",
                        
                        # Kata sapaan informal
                        "abang", "aku", "lo", "iya",
                        
                        # Kata sifat/ekspresi negatif yang tidak informatif
                        "bodoh", "goblok", "parah", "mampus",
                        
                        # Kata kerja sangat umum
                        "buat", "make", "bikin", "jadi", "terus", "masuk",
                        
                        # Interjeksi dan kata seru
                        "eh", "gas", "yuk", "sok", "banget", "gede",
                        
                        # Kata teknis media sosial
                        "rt", "via", "id", "pm", "cok", "ki", "way",}
stop_words.update(additional_stopwords)


def remove_stopwords_from_text(text):
    """
    Hapus stopwords (kecuali negasi) dari string kalimat dan kembalikan string baru
    """
    if not isinstance(text, str):
        return ""

    tokens = text.split()  # tokenize manual
    filtered_tokens = [token for token in tokens if token.lower().strip() not in stop_words]
    return " ".join(filtered_tokens)



def remove_stopwords_tokens(tokens):
    """
    Remove stopwords (kecuali negasi) dari list token
    """
    if not isinstance(tokens, list):
        return []
    
    filtered_tokens = [token for token in tokens if str(token).lower().strip() not in stop_words]
    return filtered_tokens

# Terapkan ke kolom normalized_text
df["normalized_text_no_stopwords"] = df["normalized_text"].apply(remove_stopwords_from_text)

df['tokens_normalized_text_no_stopwords'] = df['token_normalized_text'].apply(remove_stopwords_tokens)
print(stop_words)
```

```python
df_tfidf = df.copy()
```

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from scipy.sparse import save_npz
import pickle, os
from tqdm import tqdm
from sklearn.model_selection import train_test_split
```

```python
# Set parameters optimized for Indonesian text and future GNN applications
MAX_FEATURES = 7000  # Adjust based on your corpus size and memory constraints
MIN_DF = 3          # Minimum document frequency (ignore terms that appear in fewer than 3 docs)
MAX_DF = 0.85         # Maximum document frequency (ignore terms in more than 80% of docs)
NGRAM_RANGE = (1, 2) # Include both unigrams and bigrams

# Function to convert token lists to strings for vectorization
def tokens_to_string(tokens):
    """Convert token list to space-separated string"""
    if isinstance(tokens, list):
        return ' '.join([str(t) for t in tokens])
    return str(tokens)
```

```python
# Split the data into train, validation, and test sets
# Adjust these ratios based on your specific needs
print("Splitting data into train/val/test sets...")
train_ratio = 0.8
val_ratio = 0.10
test_ratio = 0.10

# First split: training and temp (validation + test)
train_df, temp_df = train_test_split(df_tfidf, test_size=(val_ratio + test_ratio), random_state=42)

# Second split: validation and test from temp
val_df, test_df = train_test_split(temp_df, test_size=test_ratio/(val_ratio + test_ratio), random_state=42)
print(f"Data split: Train: {len(train_df)} | Validation: {len(val_df)} | Test: {len(test_df)}")
```

```python
# Initialize the TF-IDF Vectorizer
print("Initializing TF-IDF vectorizer...")
tfidf_vectorizer = TfidfVectorizer(
    max_features=MAX_FEATURES,
    min_df=MIN_DF,
    max_df=MAX_DF,
    ngram_range=NGRAM_RANGE,
    sublinear_tf=True,     # Apply sublinear scaling (log scaling) - better for GNN
    norm='l2',             # L2 normalization for cosine similarity
    smooth_idf=True,       # Add 1 to document frequencies to prevent division by zero
    strip_accents='unicode',
    analyzer='word'
)
# Tentukan kolom fitur
feature_column = 'normalized_text'
```

```python
# Prepare ONLY training and validation data for fitting TF-IDF
# Important: We fit ONLY on training data and apply to validation separately
train_corpus = train_df[feature_column].apply(tokens_to_string).fillna('').values

# Fit the vectorizer ONLY on training data
print(f"Fitting TF-IDF vectorizer on {len(train_corpus)} training documents...")
tfidf_vectorizer.fit(train_corpus)

# Transform training data
train_tfidf_matrix = tfidf_vectorizer.transform(train_corpus)

# Transform validation data using the fitted vectorizer (no fitting on validation data)
val_corpus = val_df[feature_column].apply(tokens_to_string).fillna('').values
val_tfidf_matrix = tfidf_vectorizer.transform(val_corpus)

# Note: do NOT transform test data here as it should remain unseen during development
# The test transformation will only be done during final evaluation
```

```python
# Get feature names
feature_names = tfidf_vectorizer.get_feature_names_out()

print(f"Train TF-IDF matrix shape: {train_tfidf_matrix.shape}")
print(f"Validation TF-IDF matrix shape: {val_tfidf_matrix.shape}")
print(f"Number of features (vocabulary size): {len(feature_names)}")

# Create a dictionary mapping terms to their indices
term_indices = {term: idx for idx, term in enumerate(feature_names)}

# Function to get TF-IDF scores for individual documents
def get_tfidf_scores(matrix, doc_idx, top_n=10):
    """Get top terms and scores for a document"""
    if doc_idx >= matrix.shape[0]:
        return []
    
    # Get document vector and convert to array
    doc_vector = matrix[doc_idx].toarray().flatten()
    
    # Get indices of top terms
    top_indices = doc_vector.argsort()[-top_n:][::-1]
    
    # Return term-score pairs
    return [(feature_names[idx], doc_vector[idx]) for idx in top_indices 
            if doc_vector[idx] > 0]  # Only include non-zero scores

# Display top terms for the first few documents from training set
print("\nTop TF-IDF terms for sample training documents:")
for i in range(min(3, train_tfidf_matrix.shape[0])):
    print(f"\nTraining Document {i}:")
    if feature_column == 'normalized_text':
        print(f"Tokens: {train_df[feature_column].iloc[i]}")
    else:
        print(f"Text: {train_df[feature_column].iloc[i][:100]}...")
        
    top_terms = get_tfidf_scores(train_tfidf_matrix, i)
    for term, score in top_terms:
        print(f"  {term}: {score:.4f}")
```

```python
# Save the TF-IDF vectorizer and matrices
save_path = "tfidf_data"
os.makedirs(save_path, exist_ok=True)

print(f"\nSaving TF-IDF data to {save_path}/...")
save_npz(f"{save_path}/train_tfidf_matrix.npz", train_tfidf_matrix)
save_npz(f"{save_path}/val_tfidf_matrix.npz", val_tfidf_matrix)

with open(f"{save_path}/tfidf_vectorizer.pkl", "wb") as f:
    pickle.dump(tfidf_vectorizer, f)
    
with open(f"{save_path}/feature_names.pkl", "wb") as f:
    pickle.dump(feature_names, f)
```

```python
# Add TF-IDF vector to DataFrames as a new column
train_df['tfidf_vector'] = list(train_tfidf_matrix)
val_df['tfidf_vector'] = list(val_tfidf_matrix)

# Save the split dataframes
train_df.to_pickle(f"{save_path}/train_df.pkl")
val_df.to_pickle(f"{save_path}/val_df.pkl")
test_df.to_pickle(f"{save_path}/test_df.pkl")
```

```python
# Function to transform test data only when needed for final evaluation
def transform_test_data():
    """
    Transform test data using the trained TF-IDF vectorizer.
    This should be called only during final evaluation.
    """
    print("Transforming test data using pre-trained TF-IDF vectorizer...")
    test_corpus = test_df[feature_column].apply(tokens_to_string).fillna('').values
    test_tfidf_matrix = tfidf_vectorizer.transform(test_corpus)
    print(f"Test TF-IDF matrix shape: {test_tfidf_matrix.shape}")
    
    # Optionally save the transformed test data
    save_npz(f"{save_path}/test_tfidf_matrix.npz", test_tfidf_matrix)
    
    return test_tfidf_matrix

# Note: Only call this function when you're ready for final evaluation
test_tfidf_matrix = transform_test_data()
```

```python
import os
import numpy as np
import pandas as pd
import torch
import torch.nn.functional as F
from torch_geometric.nn import GCNConv
from torch_geometric.data import Data
from sklearn.metrics import f1_score, confusion_matrix, classification_report
from scipy.sparse import load_npz, vstack
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import time
from datetime import timedelta
import logging

# Set seed and device
torch.manual_seed(42)
np.random.seed(42)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# Load datasets and TF-IDF data
save_path = "tfidf_data"
train_df = pd.read_pickle(f"{save_path}/train_df.pkl")
val_df = pd.read_pickle(f"{save_path}/val_df.pkl")
test_df = pd.read_pickle(f"{save_path}/test_df.pkl")
train_tfidf_matrix = load_npz(f"{save_path}/train_tfidf_matrix.npz")
val_tfidf_matrix = load_npz(f"{save_path}/val_tfidf_matrix.npz")

with open(f"{save_path}/tfidf_vectorizer.pkl", "rb") as f:
    tfidf_vectorizer = pickle.load(f)

combined_df = pd.concat([train_df, val_df], ignore_index=True)
label_map = {'negative': 0, 'neutral': 1, 'positive': 2}
for df in [train_df, val_df, test_df, combined_df]:
    df['sentiment_label'] = df['sentiment_label_manual'].map(label_map)

combined_tfidf_matrix = vstack([train_tfidf_matrix, val_tfidf_matrix])

# Prepare ALL data (train + val + test) untuk graph
all_df = pd.concat([train_df, val_df, test_df], ignore_index=True)
all_tfidf_matrix = vstack([train_tfidf_matrix, val_tfidf_matrix, 
                          tfidf_vectorizer.transform(test_df['normalized_text'].apply(lambda x: ' '.join(x) if isinstance(x, list) else str(x)).fillna('').values)])

# Create labels for all data
all_labels = torch.LongTensor(all_df['sentiment_label'].values)

# Create graph
from sklearn.metrics.pairwise import cosine_similarity

def create_balanced_document_graph(tfidf_matrix, labels, threshold=0.05, max_neighbors=1000, cross_class_boost=1.0):
    edge_index, edge_weight = [], []
    n_docs = tfidf_matrix.shape[0]
    for i in tqdm(range(0, n_docs, 100), desc="Creating graph"):
        sims = cosine_similarity(tfidf_matrix[i:i+100], tfidf_matrix)
        for b_idx, sim in enumerate(sims):
            doc_idx = i + b_idx
            sim[doc_idx] = 0
            sim[labels != labels[doc_idx]] *= cross_class_boost
            topk = np.argsort(-sim)
            topk = [j for j in topk if sim[j] > threshold][:max_neighbors]
            for j in topk:
                edge_index.append([doc_idx, j])
                edge_weight.append(sim[j])
                edge_index.append([j, doc_idx])
                edge_weight.append(sim[j])
    return np.array(edge_index).T, np.array(edge_weight)

print("Creating document graph for ALL data...")
edge_index, edge_weight = create_balanced_document_graph(
    all_tfidf_matrix,
    all_labels.numpy()
)

# PyG data untuk ALL data
features = torch.FloatTensor(all_tfidf_matrix.toarray())
edge_index = torch.LongTensor(edge_index)
edge_weight = torch.FloatTensor(edge_weight)
data = Data(x=features, edge_index=edge_index, edge_attr=edge_weight, y=all_labels).to(device)

# Train-val-test masks
train_mask = torch.zeros(len(all_df), dtype=torch.bool)
val_mask = torch.zeros(len(all_df), dtype=torch.bool)
test_mask = torch.zeros(len(all_df), dtype=torch.bool)

train_mask[:len(train_df)] = True
val_mask[len(train_df):len(train_df)+len(val_df)] = True
test_mask[len(train_df)+len(val_df):] = True

data.train_mask = train_mask.to(device)
data.val_mask = val_mask.to(device)
data.test_mask = test_mask.to(device)

# GNN model dengan hyperparameters yang dapat di-tune
def get_model(in_features, hidden_dim, out_classes, num_layers=3, dropout=0.15):
    class GNN(torch.nn.Module):
        def __init__(self):
            super().__init__()
            self.num_layers = num_layers
            self.dropout = dropout
            
            self.reduce = torch.nn.Linear(in_features, hidden_dim)
            
            # GCN layers
            self.convs = torch.nn.ModuleList()
            for i in range(num_layers):
                self.convs.append(GCNConv(hidden_dim, hidden_dim))
            
            # FC layers
            self.fc1 = torch.nn.Linear(hidden_dim, hidden_dim)
            self.fc2 = torch.nn.Linear(hidden_dim, hidden_dim // 2)
            self.out = torch.nn.Linear(hidden_dim // 2, out_classes)
            
            # Batch norm
            self.bn = torch.nn.BatchNorm1d(hidden_dim)
            self.bn2 = torch.nn.BatchNorm1d(hidden_dim // 2)
            
        def forward(self, x, edge_index, edge_weight):
            x = F.relu(self.bn(self.reduce(x)))
            x = F.dropout(x, self.dropout, self.training)
            
            # GCN layers with residual connections
            for i, conv in enumerate(self.convs):
                res = x
                x = F.relu(conv(x, edge_index, edge_weight))
                if i > 0:  # Skip connection from second layer
                    x = x + res
                x = F.dropout(x, self.dropout, self.training)
            
            x = F.relu(self.fc1(x))
            x = F.relu(self.bn2(self.fc2(x)))
            return self.out(x)
    return GNN()

optimal_params = {
    'hidden_dim': 64,
    'num_layers': 5,
    'dropout': 0.4,
    'lr': 0.00425,
    'weight_decay': 1.07e-07, # 1.85e-05 FOR HIGHER ACC, BUT NOT CONSTANT 
    'label_smoothing': 0.109,
    'neg_weight': 1.0,
    'neu_weight': 1.1,
    'pos_weight': 1.1,
    'T_0': 10,
    'T_mult': 2,
    'eta_min': 4.98e-07
}

print("Using optimal hyperparameters:")
for key, value in optimal_params.items():
    print(f"  {key}: {value}")

# Create model with optimal parameters
model = get_model(
    data.num_features, 
    optimal_params['hidden_dim'], 
    3, 
    optimal_params['num_layers'], 
    optimal_params['dropout']
).to(device)

optimizer = torch.optim.AdamW(
    model.parameters(), 
    lr=optimal_params['lr'], 
    weight_decay=optimal_params['weight_decay']
)

scheduler = torch.optim.lr_scheduler.CosineAnnealingWarmRestarts(
    optimizer, 
    T_0=optimal_params['T_0'], 
    T_mult=optimal_params['T_mult'], 
    eta_min=optimal_params['eta_min']
)

class_weights = torch.tensor([
    optimal_params['neg_weight'], 
    optimal_params['neu_weight'], 
    optimal_params['pos_weight']
]).to(device)

print("\nStarting training with optimal parameters...")

# Training dengan optimal parameters
train_losses, val_losses, val_accs, val_f1s = [], [], [], []
best_val_f1 = 0
best_epoch = 0
best_model_state = None

start_time = time.time()

for epoch in range(150):
    model.train()
    optimizer.zero_grad()
    out = model(data.x, data.edge_index, data.edge_attr)
    loss = F.cross_entropy(
        out[data.train_mask], 
        data.y[data.train_mask], 
        weight=class_weights,
        label_smoothing=optimal_params['label_smoothing']
    )
    loss.backward()
    torch.nn.utils.clip_grad_norm_(model.parameters(), 0.775) #0.775
    optimizer.step()
    scheduler.step()

    # Validation
    model.eval()
    with torch.no_grad():
        out = model(data.x, data.edge_index, None)
        probs = F.softmax(out, dim=1)
        preds = torch.argmax(probs, dim=1)
        
        val_preds = preds[data.val_mask]
        val_labels = data.y[data.val_mask]
        
        acc = (val_preds == val_labels).float().mean().item()
        f1 = f1_score(val_labels.cpu(), val_preds.cpu(), average='weighted', zero_division=0)
        
        val_ce_loss = F.cross_entropy(out[data.val_mask], val_labels)

    train_losses.append(loss.item())
    val_losses.append(val_ce_loss.item())
    val_accs.append(acc)
    val_f1s.append(f1)
    
    if f1 > best_val_f1:
        best_val_f1 = f1
        best_epoch = epoch + 1
        best_model_state = model.state_dict().copy()

    if epoch % 10 == 0:
        print(f"Epoch {epoch+1:03d}: TrainLoss={loss.item():.4f} ValLoss={val_ce_loss.item():.4f} ValAcc={acc:.4f} ValF1={f1:.4f}")

training_time = time.time() - start_time

# Load best model dan evaluasi final
if best_model_state is not None:
    model.load_state_dict(best_model_state)

model.eval()
with torch.no_grad():
    # Gunakan forward dengan graph structure
    out = model(data.x, data.edge_index, data.edge_attr)
    probs = F.softmax(out, dim=1)
    preds = torch.argmax(probs, dim=1)
    
    # Ambil hasil untuk test set
    test_preds = preds[data.test_mask]
    test_labels_actual = data.y[data.test_mask]

test_acc = (test_preds == test_labels_actual).float().mean().item()
test_f1 = f1_score(test_labels_actual.cpu(), test_preds.cpu(), average='weighted', zero_division=0)

print(f"\nFinal Results:")
print(f"Training Time: {timedelta(seconds=training_time)}")
print(f"Best Validation F1: {best_val_f1:.4f}")
print(f"Test Accuracy: {test_acc:.4f}")
print(f"Test F1 Score: {test_f1:.4f}")
print(f"Best Epoch: {best_epoch:.4f}")

# Classification report
print("\nClassification Report:")
report = classification_report(
    test_labels_actual.cpu(), 
    test_preds.cpu(), 
    target_names=['Negative', 'Neutral', 'Positive'],
    zero_division=0
)
print(report)

# Confusion Matrix
cm = confusion_matrix(test_labels_actual.cpu(), test_preds.cpu())
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Neg', 'Neu', 'Pos'],
            yticklabels=['Neg', 'Neu', 'Pos'])
plt.title(f'Optimized GCN Confusion Matrix (Test Acc: {test_acc:.4f})')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.savefig('optimal_gcn_confusion_matrix.png')
plt.show()

# Training curves
plt.figure(figsize=(12, 8))
plt.subplot(2, 1, 1)
plt.plot(train_losses, label='Training Loss')
plt.plot(val_losses, label='Validation Loss')
plt.title('Training Loss Curve - Optimal GCN')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid(True)
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(val_accs, label='Validation Accuracy')
plt.plot(val_f1s, label='Validation F1 Score')
plt.axhline(y=test_acc, color='r', linestyle='--', label=f'Test Accuracy: {test_acc:.4f}')
plt.axhline(y=test_f1, color='g', linestyle='--', label=f'Test F1 Score: {test_f1:.4f}')
plt.title('Validation Metrics - Optimal GCN')
plt.xlabel('Epoch')
plt.ylabel('Score')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.savefig('optimal_gcn_training_curves.png')
plt.show()

# Save results
test_probs_final = probs[data.test_mask]
results_df = pd.DataFrame({
    'text': test_df['text_clean'] if 'text_clean' in test_df else test_df['text'],
    'true': test_df['sentiment_label_manual'],
    'pred': ['negative' if p == 0 else 'neutral' if p == 1 else 'positive' for p in test_preds.cpu().numpy()],
    'confidence': torch.max(test_probs_final, dim=1)[0].detach().cpu().numpy()
})
results_df.to_csv('optimal_gcn_test_predictions.csv', index=False)

# Summary
num_parameters = sum(p.numel() for p in model.parameters() if p.requires_grad)
print("\n========== Optimal GCN Model Summary ==========")
print(f"Number of Parameters: {num_parameters:,}")
print(f"Training Time: {timedelta(seconds=training_time)}")
print(f"Best Validation F1: {best_val_f1:.4f}")
print(f"Test Accuracy: {test_acc:.4f}")
print(f"Test F1 Score: {test_f1:.4f}")
print(f"Best Epoch: {best_epoch:.4f}")
print("================================================")

# Save model
torch.save({
    'model_state_dict': best_model_state if best_model_state is not None else model.state_dict(),
    'optimal_params': optimal_params,
    'test_acc': test_acc,
    'test_f1': test_f1,
    'best_val_f1': best_val_f1
}, 'optimal_gcn_model.pth')
print("Model saved to 'optimal_gcn_model.pth'")
```

```python
from torch_geometric.utils import degree

def show_graph_statistics(data):
    num_nodes = data.num_nodes
    num_edges = data.edge_index.shape[1] // 2  # assume undirected, edges are duplicated

    deg = degree(data.edge_index[0], num_nodes=num_nodes)
    mean_deg = deg.mean().item()
    max_deg = deg.max().item()
    min_deg = deg.min().item()

    print("\n====== Graph Statistics ======")
    print(f"Number of Nodes       : {num_nodes}")
    print(f"Number of Edges       : {num_edges}")
    print(f"Average Degree        : {mean_deg:.2f}")
    print(f"Max Degree            : {max_deg}")
    print(f"Min Degree            : {min_deg}")
    print("================================")

    # Optional: visualize degree distribution
    plt.figure(figsize=(8, 4))
    plt.hist(deg.cpu().numpy(), bins=50, color='skyblue', edgecolor='black')
    plt.title("Analisis Distribusi Derajat Node Tanpa Undersampling")
    plt.xlabel("Derajat")
    plt.ylabel("Jumlah")
    plt.grid(True)
    plt.tight_layout()
    plt.savefig('node_degree_distribution.png')
    plt.show()

show_graph_statistics(data)
```
